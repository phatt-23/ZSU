# Encode the categorical data

- We can either encode the categorical features with an OrdinalEncoder or OneHotEncoder
- OrdinalEncoder is good for categories with an ordering
- OneHotEncoder is good for categories without a clear ordering
---
- Therefore I choose one hot encoding, because non of the categorical features have a clear ordering

````python
# Making a convenience function to encode columns

def encode_columns_one_hot(df: pd.DataFrame, columns: list[str]):
    for column in columns:
        encoded_column = pd.get_dummies(df[column], prefix=column, dtype=int)
        df = pd.concat([df, encoded_column], axis=1)
        df = df.drop(column, axis=1)
    return df

def encode_columns_ordinal(df: pd.DataFrame, columns: list[str]):
    for column in columns:
        categories = list(df[column].unique())

        encoder = OrdinalEncoder(categories=[categories])
        encoded_column = encoder.fit_transform(df[[column]])

        df.loc[:, column] = encoded_column
    return df
````

````python

# Encoding all categorical features with via new function

# extract the names of the categorical features
categorical_feature_names = list(df.describe(exclude=np.number).columns)

# encode the features
df = encode_columns_ordinal(df, categorical_feature_names)

# display the results
print(f"{categorical_feature_names = }")

display(df.shape)
display(df.head())
````

# Choosing a splitter

- KFold is good for balanced data
- StratifiedKFold is good when data is imbalanced


````python
df.Target.value_counts()
````

Our data is very imbalanced, so we'll use StratifiedKFold.

````python
kfold = StratifiedKFold(n_splits=5)
splits = kfold.split(X_train, y_train)
score = []

for train_index, test_index in splits:
    X_train_fold, X_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

    clf = DecisionTreeClassifier(random_state=13)
    clf.fit(X_train_fold, y_train_fold)

    y_pred_fold = clf.predict(X_test_fold)
    score.append(f1_score(y_test_fold, y_pred_fold))

np.mean(score), np.min(score), np.max(score)
````

I wrote a generic function that makes training the splits less of an eye sore.

````python
def train_splits(model, model_selection, X_data, y_data):
    """
    model = training model
    model_selection = kfolder
    X = source
    y = target
    """

    splits = model_selection.split(X_data, y_data)
    score = []

    for train_index, test_index in splits:
        X_train_fold, X_test_fold = X_data.iloc[train_index], X_data.iloc[test_index]
        y_train_fold, y_test_fold = y_data.iloc[train_index], y_data.iloc[test_index]

        model.fit(X_train_fold, y_train_fold)
        y_pred_fold = model.predict(X_test_fold)

        score.append(f1_score(y_test_fold, y_pred_fold))

    return { 'mean': np.mean(score), 'min': np.min(score), 'max': np.max(score) }
````

````python
default_models = [
    DecisionTreeClassifier(random_state=0),
    RandomForestClassifier(random_state=0),
]

results = []

for model in default_models:

    result = train_splits(
        model           = model,
        model_selection = StratifiedKFold(n_splits=5),
        X_data          = X_train,
        y_data          = y_train
    )

    y_pred = model.predict(X_test)
    score = f1_score(y_test, y_pred)

    results.append({
        'model': model,
        'result': result,
        'score': f1_score(y_test, y_pred),
    })

for result in results:
    display(result)
````

# Hyper-parameter Optimization

````python
from sklearn.model_selection import GridSearchCV

# ML algorithms
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# hyper-parameter optimizers
from sklearn.experimental import enable_halving_search_cv  # noqa
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import HalvingRandomSearchCV
from sklearn.model_selection import HalvingGridSearchCV

# tools
from scipy.stats import uniform
````

Writing a convenience function.

````python
def find_optimal_classifier(
    hyper_param_searcher_klass,
    classifier_model_klass,
    cross_validator,
    hyper_param_distributions,
    X, y,
):
    classifier_model = classifier_model_klass(random_state=0)
    hyper_param_searcher = hyper_param_searcher_klass(
        estimator   = classifier_model,
        param_grid  = hyper_param_distributions,
        cv          = cross_validator,
        scoring     = 'f1'
    )

    optimal_classifier = hyper_param_searcher.fit(X, y)
    return optimal_classifier


def search_optimal_classifiers(
    searcher_klass,
    cross_validator,
    model_hyper_parameters,
):
    optimal_classifiers = []

    for model, hyper_param_distributions in model_hyper_parameters.items():
        search = find_optimal_classifier(
            hyper_param_searcher_klass  = searcher_klass,
            classifier_model_klass      = model,
            cross_validator             = cross_validator,
            hyper_param_distributions   = hyper_param_distributions,
            X                           = X_train,
            y                           = y_train,
        )

        df_result = pd.DataFrame(search.cv_results_)
        df_result = df_result[['rank_test_score', 'mean_test_score'] + [f'split{i}_test_score' for i in range(0, cross_validator.n_splits)]]
        df_result = df_result.sort_values(by='rank_test_score')

        optimal_classifiers.append({
            'estimator': search.best_estimator_,
            'results': df_result,
        })

    return optimal_classifiers
````

Define the search space for the models that will be used in the search for the best classifiers.

Search for the optimal classifiers - train them on the training set.

Then try out the trained model on test data.

````python
# defining models and their hyper parameter grid
# for the hyper-parameter optimizers to optimize

model_hyper_parameters = {
    DecisionTreeClassifier: {
        'min_samples_split' : range(2,20),
        'max_depth'         : range(5,10),
    },
    RandomForestClassifier: {
        'n_estimators'      : range(10,150,10),
        'max_depth'         : range(5,10),
    },
}

# learning on training dataset

optimal_classifiers = search_optimal_classifiers(
    searcher_klass          = GridSearchCV,
    cross_validator         = StratifiedKFold(n_splits=5),
    model_hyper_parameters  = model_hyper_parameters,
)

# displaying the results

for classifier in optimal_classifiers:
    display(classifier['results'])
    display(classifier['estimator'])
````

We only trained on the training data. Now try out the trained model on test data.

````python
# predicting test dataset
scores = []

for classifier in optimal_classifiers:
    estimator = classifier['estimator']
    y_pred = estimator.predict(X_test)

    scores.append({
        'estimator': estimator,
        'score': f1_score(y_test, y_pred),
    })

scores
````
